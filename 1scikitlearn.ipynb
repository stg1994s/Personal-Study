## 가상 데이터 생성하기
### 지금까지는 이미지를 분석한다든지, 회귀분석(보스턴 집값 예측, iris는 분류였고.)
### 하지만 이제 분류 테스트를 위해서 make_blob, make_classification.
###    make_moon, make_circle등을 이용해서 가상의 데이터를 생성할 수 있음.
###      따라서 분류나 군집이나 이런것들 할 수 있을것임.

```python
# make_classification

from sklearn.datasets import make_classification
```

```python
X, y = make_classification(
    # n_samples : 표본수, 기본 : 100
    n_features=2, # 독립변수 수 
    n_informative=1, # x,y의 상관관계여부
    n_redundant=0,  # x,y의 선형관계여부
    n_clusters_per_class=1, # 군집수
    random_state=200781430)

plt.scatter(X[:,0], X[:, 1], marker='o', 
            c=y, s=100, edgecolor='k')
```

### 이것을 통해서 임의의 데이터 배열을 만들 수 있다는 겁니다.
### 표본 수를 임의로 증가시키지 않으면, 기본은 100개, 이정도로 생각하시면 될 것 같습니다.

```python
# 정규분포를 통해 군집용 가상데이터 생성

from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_features=2, 
                  centers = 3, # 군집 중싱점 갯수. 센터르 중심으로 점들이 얼마나 많이 몰리느냐, 를 파악.
                  random_state=2111031705)
plt.scatter(X[:, 0], X[:,1], marker='o', c=y, s=100, edgecolors='k')       
```
