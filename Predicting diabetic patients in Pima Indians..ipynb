### 피마 인디언의 당뇨병 예측 실행
 + 이제 케라스를 이용해 당뇨병 예측을 실행하면 다음과 같다. 먼저 seed 값 설정을 먼저 알아보자.

```python
## 필요한 라이브러리 불러오기

from keras.models import Sequential
from keras.layers import Dense
import numpy
import tensorflow as tf
```

```python
# seed 값 생성
seed = 0
numpy.random.seed(seed)
tf.random.set_seed(seed)
```

#### Random()함수를 써서 임의의 숫자를 만들어 내는 것처럼 보여도, 
 + 이는 컴퓨터 안에 미리 내장된 수많은 '랜덤 테이블' 중 하나를 불러내 그 표의 순서대로 숫자를 보여주는 것.
 + 따라서 SEED 값을 설정한다는 것은 그 랜덤 테이블 중에서 몇 번째 테이블을 불러와 쓸지를 정하는 것.
 + 따라서 SEED 값이 같으면 똑같은 랜덤 값을 출력한다.

```python
# 데이터 로드
dataset = numpy.loadtxt("C:/Users/korea/Desktop/My Jupyter/pima-indians-diabetes.csv", delimiter=",")
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
# 모델의 설정
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# 모델 컴파일
model.compile(loss='binary_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
```

### 컴파일 이후 모델을 실행하기 전에, 이 예제는 둘중 하나를 결정하는 이항 분류(Binary Classification) 문제이다.
 + 그러므로 오차 함수는 binary_crossentropy를 사용하고, 최적화 함수로 adam을 사용한다.
 + 그리고 전체 샘플이 200번 반복해서 입력될 때까지 실험을 반복한다. 한번에 입력되는 입력 값을 10개로 하고,
 + 종합하면 다음과 같은 프로그램이 완성된다.

```python
# 모델 실행
model.fit(X, Y, epochs=200, batch_size=10)
```

```python
# 결과 출력
print("\n Accuracy: %.4f" % (model.evaluate(X, Y)[1]))
```
