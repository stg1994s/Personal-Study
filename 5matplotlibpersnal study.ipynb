```python
np.random.seed(2111021105)
y = np.random.standard_normal(20) # 정규분포 난수 20개
x = np.arange(1, 20+1)
plt.plot(y)
```

```python
plt.plot(x,y)
```

```python
# 암시적 스타일 지정 : 색
# 색 : r g b c m y k w
plt.plot(x,y, 'r')
```

```python
# 암시적 스타일 지정 : 마커
# 데이터 위치를 나타내는 기호
# . o v ^ 1 p * D
plt.plot(x,y, 'D')
```

```python
plt.plot(x,y, 'rD')
```

```python
# 암시적 스타일 지정 : 선모양
# 실선, 대시선, 점선, 대시점선
# - : -- -.
```

```python
plt.plot(x,y, 'r-.')
```

```python
# 명시적 스타일 지정 
# color. linestyle, linewidth
# markersize, markerfacecolor, markereagecolor
plt.plot(x,y, c='m', ls='-.', lw=3, marker='d', ms=10, mfc='k', mec='k') 
# color='m', linestyle='-.', linewidth=3, marker='d', markersize=10, markercolor='black', markereagecolor='black'
```

```python
# 축 지정하기
# xlim(최소, 최대), ylim(최소, 최대) 으로 설정
plt.plot(x,y)
plt.xlim(-1, 22)
plt.ylim(-3, 3)
```

```python
popmove = pd.read_excel('시도별 전출입 인구수.xlsx')
popmove.head()
```

```python
# 결측치를 0 으로 채움
popmove.fillna(0) # 의미없음
```

```python
# 결측치를 이전행의 값으로 대체
popmove = popmove.fillna(method='ffill')
```

```python
# 데이터 추출
serach = (popmove.전출지별 == '전국') & (popmove.전입지별 == '서울특별시')
```

```python
seoulmove = popmove[serach]
seoulmove
```

```python
x = np.arange(2000,2017+1)
y = seoulmove.iloc[:,32:]
y # wide 하게 출력 
```

```python
y = y.iloc[0, ]
y # long 하게 출력
```

```python
plt.plot(x,y)
```

```python
# 축 눈금 지정하기
# 축 상의 눈금 tick이라 함
# xticks(수치리스트, 레이블)
# yicks(수치리스트, 레이블)
```

```python
# linspace : 지정한 범위를 특정 수만큼 일정하게 나눔
np.linspace(1,10, 10)
np.linspace(0,1, 10)
```

```python
np.pi
x = np.linspace(-np.pi, np.pi, 256)
y = np.cos(x)
plt.plot(x,y)
# y축 눈금이 자세히 출력됨
```

```python
plt.plot(x,y)
plt.xticks([-3.14,-3.14/2,0,3.14/2,3.14],['a','b','c','d','e'])
plt.yticks([-1,-0.5,0,0.5,1],['가','나','다','라','마'])
plt.show()
```

```python
# 안내선 출력 : grid line
# tick의 위치를 잘 보여주기 위해 사용
# grid(True/False)
plt.plot(x,y)
plt.grid(True)
```

```python
# 범례 표시
# 여러 개의 plot으로 그래프를 그린 경우
# 각 plot이 어떤 자료를 나타내는지 알려야 함
# 자동으로 범례를 표시하려면 각 그래픝 작성시
# label로 이름을 지정해야 함
y1 = np.cos(x)
y2 = np.sin(x)

plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.plot(x,y1, 'r--', label='코사인')
plt.plot(x,y2, 'b--', label='사인')
plt.legend(loc='best')
```
