```python
# 간단한 sin 그래프
x = np.linspace(0,9.5,1000)
```

```python
fig = plt.figure()
ax = plt.axes() # 축영역
ax.plot(x,np.sin(x), 'r')
ax.grid()
```

```python
fig = plt.figure()
ax = plt.axes() # 축영역
ax.plot(x,np.cos(x), 'b')
ax.grid()
```

```python
fig = plt.figure()
ax = plt.axes() # 축영역
ax.plot(x,np.tan(x/2), 'g')
ax.set_xlim(2,4)
ax.set_ylim(50,-50)
ax.grid()
```

```python
# sin cos, tan 같이 그리기 
fig = plt.figure()
ax = plt.axes() # 축영역
ax.plot(x,np.sin(x), 'r', label = '사인')
ax.plot(x,np.cos(x), 'b', label = '코사인')
ax.plot(x,np.tan(x), 'g', label = '탄젠트')
ax.set(xlim=(0,9.5),ylim=(-1,1), title='sim/cos/tan 그래프', xlabel='x값',ylabel='y값')
ax.legend()
ax.grid()
```

```python
# 여러개의 그래프 나타내기 : subplot
# subplot(행수/열수/번호)

x1 = np.linspace(0,5)
x2 = np.linspace(0,2)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2) 
```

```python
# 2X1 그래프 
fig = plt.figure()
ax1 = plt.subplot(221)
plt.plot(x1, y1, 'ro--')

ax2 = plt.subplot(212)
plt.plot(x2,y2, 'b*:')
```

```python
# 1X2 그래프 
fig = plt.figure()
ax1 = plt.subplot(121)
plt.plot(x1,y1, 'ro--')

ax2 = plt.subplot(122)
plt.plot(x2,y2, 'b*:')
```

```python
# 부산 -> 서울, 경기, 인천 인구 이동을 시각화해보세요
popmove = pd.read_excel('시도별 전출입 인구수.xlsx', header=0)
```

```python
popmove = popmove.fillna(method='ffill')
```

```python
search = (popmove.전출지별 == '부산광역시') & (popmove.전입지별 == '서울특별시')
```

```python
psmove = popmove[search]
psmove
```

```python
# 부산 -> 경기
search = (popmove.전출지별 == '부산광역시') & (popmove.전입지별 == '경기도')
pkmove = popmove[search]
pkmove
```

```python
# 부산 -> 인천
search = (popmove.전출지별 == '부산광역시') & (popmove.전입지별 == '인천광역시')
pimove = popmove[search]
pimove
```

```python
# 변수 생성
x = np.arange(2000, 2017+1)
y1 = psmove.iloc[:, 32:].iloc[0]
y2 = pkmove.iloc[:, 32:].iloc[0]
y3 = pimove.iloc[:, 32:].iloc[0]
```

```python
# 그래프로 시각화
plt.plot(x,y1, 'r', label='부산-서울')
plt.plot(x,y2, 'b', label='부산-경기')
plt.plot(x,y3, 'm', label='부산-인천')
plt.legend()
```

```python
# 그래프로 시각화2
fig = plt.figure()
ax = plt.axes()
plt.plot(x,y1, 'r', label='부산-서울')
plt.plot(x,y2, 'b', label='부산-경기')
plt.plot(x,y3, 'm', label='부산-인천')
plt.legend()
# x/y축 눈금을 리스트로 변환후 포맷문자열로 변환
ax.set_xticklabels(['{:}'.format(int(x))for x in ax.get_xticks().tolist()])
ax.set_yticklabels(['{:}'.format(int(y))for y in ax.get_xticks().tolist()])
```

```python
# 그래프로 시각화3
plt.subplot(2,1,1)
plt.plot(x,y1,'r')
plt.title('부산->서울')
plt.xlim([2000,2017])
plt.show()
```

```python
plt.subplot(2,2,4, sharey=ax2) # 이전 그래프의 y축을 공유
plt.plot(x,y2,'m')
plt.title('부산->인천')
plt.xlim(2000,2017)
plt.yticks(visible=False) # 눈금 표시 안함
plt.tight_layout()
plt.show()
```
