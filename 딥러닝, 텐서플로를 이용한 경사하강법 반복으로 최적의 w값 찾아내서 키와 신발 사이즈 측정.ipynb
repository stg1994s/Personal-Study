# 딥러닝으로 수학 문제 풀어보자.

```python
import tensorflow as tf
import keras
```

```python
키 = [170,180,165,160]
신발 = [260, 270, 265, 255]
```

```python
# 문제 1 -> 키로 신발사이즈를 추정해보자.

키 = 170
신발 = 260

a = tf.Variable(0.1)
b = tf.Variable(0.2)


def 손실함수():
    예측값 = 키 * a + b
    return tf.square(260 - 예측값) # return을 이용해서 손실값을 뱉어내준다는 개념. 손실값 계산법은 실제값 - 예측값


# 이 변수를 학습시킵시다. 딥러닝을 이용해서.

opt = tf.keras.optimizers.Adam(learning_rate=0.1)

# 원하는 만큼 경사하강 반복하는 for문 코드

for i in range(300): 
    opt.minimize(손실함수, var_list=[a,b]) 
    print(a.numpy(),b.numpy())
                     
    
# 경사하강 한번 해줌.
# 경사하강법을 이용해서 지가 알아서 업데이트 해주는 고마운 도구입니다. 
# 이것만 있으면 경사하강법으로 w값 업데이트가 복잡한 수식 계산 없이 자동으로 일어남. 이것이 텐서플로우의 장점.
```

```python
# 맨 마지막에 나온 값이 경사하강법으로 최적의 w 값(가중치)를 찾아낸 것. 키에 저 두 값을 대입해보자. 키 * 1.52 + 1.62

170 * 1.52 + 1.62
```

```python
# 자, 답이 나왔지? 이것이 바로 선형회귀 모델이라는 것이다. 저 키 값에 다른 키를 대입하면 적절한 발사이즈가 나올 것이다. 
# 한번 해보자.

180 * 1.52 + 1.62
```
